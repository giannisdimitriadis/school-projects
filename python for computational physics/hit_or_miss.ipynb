```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def U() : return np.random.rand() #Γεννήτρια τυχαίων αριθμών
def f(x) : return (np.exp(x)-1)/(np.exp(5)-1) 
```

```python
N = 100000
xArray = []
for i in range(N):
    #Χρειαζόμαστε δύο τυχαίους αριθμούς. Ο ένας θα τρέχει στο [0,5], και ο άλλος θα τρέχει απ'το 0 ως το max της f, δηλαδή στο [0,f(5)]
    x = 5*U() #[0,5]
    y = f(5)*U() #[0,e^5]
    if y < f(x) :
        xArray += [x]
print('accepted x = ', len(xArray))
```

```python
plt.hist(xArray, bins = 50)
```

```python
#Το θετικό της μεθόδου απόρριψης είναι ότι δεν χρησιμοποιώ πουθενά την αντίστορφη συνάρτηση, παρά μόνο την ίδια την συνάρτηση.
#Το αρνητικό είναι ότι πρέπει να γεννήσω μεγαλύτερο πλήθος αριθμών. Δηλαδή εδώ έχω γεννήσει 100.000 x και 100.000 y για να πάρω μόλις 20.000 τυχαία νούμερα που κατανέμονται με βάση την πυκνότητα πιθανότητας e^x, που θέλαμε. Αυτό αν θέλω μπορώ να το δω αν πλοτάρω το σκατερ των σημείων x,y και σκεφτώ πόσα από αυτά τα σημεία βρίσκονται κάτω απ'την εκθετική συνάρτηση σε αυτό το διάστημα.
N = 100
xArray = []
xcoord = []
ycoord = []
for i in range(N):
    x = 5*U() #[0,5]
    y = f(5)*U() #[0,e^5]
    xcoord += [x]
    ycoord += [y]
    if y < f(x) :
        xArray += [x]
print('accepted x = ', len(xArray))
plt.scatter(xcoord, ycoord) 
```
