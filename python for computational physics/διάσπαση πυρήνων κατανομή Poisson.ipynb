```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pdecay():
    res = False
    u = np.random.rand()
    if u < 10**(-3): res = True         # Συνθήκη διάσπασης (αν ο αριθμός που γεννάω είναι μικρότερος του 0.001 τότε θεωρώ ότι ο εκάστοτε πυρήνας διασπάται)
    return res
```

```python
N = 30000              # Αρχικοποίηση αδιάσπαστων πυρήνων
nc = []               # Πυρήνες που είναι αδιάσπαστοι την εκάστοτε χρονική στιγμή
ΔN = []               # Ο αριθμός των πυρήνων που διασπάται στο βήμα Δt (αυτό που καταγράφει ο ανιχνευτής)

while (N > 0):        # Κάθε loop είναι πρακτικά ένα χρονικό στιγμιότυπο Δt
    ndecays = 0
    for i in range(N):
        if pdecay(): ndecays += 1   
    N = N - ndecays
    ΔN += [ndecays]
    nc += [N]

# Εφόσον όλο το while loop αντιπροσωπεύει ένα χρονικό στιγμιότυπο, γίνεται εμφανές ότι, εφόσον για κάθε πυρήνα ξεχωριστά ελέγχεται αν υφίσταται διάσπαση, μπορούμε σε κάθε χρονικό βήμα να έχουμε περισσότερους από έναν πυρήνες που διασπώνται
```

```python
x = [i for i in range(len(nc))]      # Είναι ο χρόνος μου διακριτοποιημένος ανά 1 δευτερόλεπτο
y = nc                               # Πυρήνες που είναι αδιάσπαστοι την εκάστοτε χρονική στιγμή
plt.plot(x, y)           # Είναι το ίδιο να μην ορίσω καν τα x, y και να πω plt.plot(nc)
```

```python
plt.hist(ΔN[2000:3000])     # Μας δίνει το νούμερο που θα έδινε ο ανιχνευτής Geiger-Muller στο εκάστοτε χρονικό βήμα. Όσο αυξάνω το Ν η κατανομή τείνει περισσότερο στην κατανομή Poisson (δεν είναι τόσο απλό, πρέπει να ναι περίπου σταθερός και ο αριθμός των πυρήνων που διασπάται ανά χρονικό βήμα). Αυτό φαίνεται καλά για Ν=30000 και βλέποντας τα ΔΝ στο διάστημα 2000 εως 3000
```
