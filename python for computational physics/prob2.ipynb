```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f=open("dices.txt", "r") #Άνοιγμα του αρχείου που περιέχει τα δεδομένα με τις ζαριές
```

```python
array = np.array(f.read().splitlines()) #Αποθήκευση των δεδομένων σε λίστα
array = array.astype(int) #Μετατροπή σε τύπο int
```

```python
array
print(len(array))
```

```python
plt.hist(array, bins = range(1,8))
```

```python
games = []
j = 0
for i in range(0,200):
    #print(j)
    ndices = 0 # Πλήθος ζαριών
    pos = 1 # Παρακολουθεί σε ποιά θέση βρίσκεται το πιόνι
    while (pos < 25):
        pos += array[j] #Η θέση του ζαριού αυξάνεται κάθε φορά κατά την ζαριά που αντλείται απ' το array που περιέχει τα δεδομένα των ζαριών
        j += 1
        ndices += 1
        if pos == 3 : pos = 24 # σκάλα
        if pos == 21 : pos = 7 # φιδάκι
    games.append(ndices)
print(games)
plt.hist(games, bins = range(0,20), histtype = 'bar', ec = 'black')
```
