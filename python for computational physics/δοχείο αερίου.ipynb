```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N1 = 1000         # Αρχικός πληθυσμός του Ν1
N2 = 0            # Αρχικός πληθυσμός του Ν2
p = 1/100         # Πιθανότητα ενός μορίου να μεταβεί στο άλλο δοχείο
T = 500           # Συνολικός αριθμός χρονικών βημάτων Δt

N1t = []          # Δημιουργία λίστας που θα αποθηκεύει τις τιμές του Ν1
N2t = []          # Δημιουργία λίστας που θα αποθηκεύει τις τιμές του Ν2
N1t += [N1]       # Εισάγουμε την αρχική τιμή του Ν1 στη λίστα Ν1t
N2t += [N2]       # Εισάγουμε την αρχική τιμή του Ν2 στη λίστα Ν2t


for i in range(0,T):     # T χρονικά βήματα Δt
    d12 = 0       # Επαναμηδενισμός μεταβλητής που μετρά πόσα μεταβαίνουν απ'το 1ο στο 2ο σε Δt
    d21 = 0       # Επαναμηδενισμός μεταβλητής που μετρά πόσα μεταβαίνουν απ'το 2ο στο 1ο σε Δt

    for j in range(0,N1):
        if (np.random.rand() < p):
            d12 += 1

    for j in range(0,N2):
        if (np.random.rand() < p):
            d21 += 1

    N1 += d21 - d12   # Προσθαφαιρούμε τα μόρια που "κέρδισε" και "έχασε" το δοχείο 1
    N2 += d12 - d21   # Προσθαφαιρούμε τα μόρια που "κέρδισε" και "έχασε" το δοχείο 2

    # Περνάμε τις νέες αυτές τιμές στις αντίστοιχες λίστες
    N1t += [N1]
    N2t += [N2]


# Σχεδιάζουμε τα διαγράμματα της χρονικής εξέλιξης των δύο πληθυσμών (Πληθυσμός 1 με κυανό και πληθυσμός 2 με μωβ)
plt.figure(figsize=(16,9))
plt.rc('font', size=22)
plt.xlabel('χρονικά βήματα Δt')
plt.ylabel('Πληθυσμός Ν1, Ν2')

plt.plot(N1t, color='cyan', linestyle='-', linewidth=4)
plt.plot(N2t, color='magenta', linestyle='-', linewidth=4)


# Ελέγχουμε αν ο τελικός αριθμ΄ςο των μορίων είναι ίδιος με τον αρχικό, για επιβεβαίωση λειτουργικότητας της μεθόδου
print("N1t[0] = ", N1t[0], "\tN2t[0] =", N2t[0], "\tsum =", N1t[0]+N2t[0],
      "N1t[T] = ", N1t[T], "\tN2t[T] =", N2t[T], "\tsum =", N1t[T]+N2t[T],
      "\nT =", T)
```
