```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def U() : return np.random.rand()                     # Γεννήτρια τυχαίων αριθμών
def Finv(x) : return np.log((np.exp(5)-1)*x+1)        # Αντίστροφη συνάρτηση της F(x)=(e^x-e^a)/(e^b-e^a) για a=0, b=5 (F^-1(x)=ln[x(e^5-1)+1])


# Η μέθοδος αυτή προϋποθέτει ότι γνωρίζουμε την αντίστροφη συνάρτηση της συνάρτησης υπό μελέτη. Γι αυτό και δεν είναι προτιμητέα.
```

```python
xArray = [Finv(U()) for i in range(0,10000)]
```

```python
plt.hist(xArray, bins = 50) # Βλέπουμε ότι εμφανίζεται το σχήμα μιας εκθετικής κατανομής (Παράγουμε δηλαδή τυχαία νούμερα, αλλά πολύ περισσότερα κοντά στο 5 απ' ότι στο 0)
```
