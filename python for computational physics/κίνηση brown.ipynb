```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step():
    l = 1
    if np.random.rand() < 0.5 : l = -1
    return l
```

```python
#     1 

N = 20000          #  
path = []          #      x   
x = 0              #   
path = [x]

for n in range(N):
    x += step()
    path += [x]

plt.plot(path)     #         
#     ,          ,      ,         
```

```python
#    = 1000  Brown    = 200  (         )

M = 1000            #  
N = 200             #  
xArray = []         #  1x1000

for m in range(M):
    x = 0
    for n in range(N):
        x += step()
    xArray += [x]
```

```python
xArray[0:10] #         10       
#xArray # M        '       

plt.hist(xArray, bins = range(-60,61)) #       
#                   (       x=0)      (200  )        '   0
```

```python
#     numpy array       ,               '  0          sqrt(N),   .

xArray = np.array(xArray)
print('mean = ', xArray.mean(), 'std = ', xArray.std(), 'sqrt(N) = ', N**0.5)
```
