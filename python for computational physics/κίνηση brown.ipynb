```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step():
    l = 1
    if np.random.rand() < 0.5 : l = -1
    return l
```

```python
# Εφαρμογή Ν βημάτων για 1 σωματίδιο

N = 20000          # Αριθμός βημάτων
path = []          # Λίστα που αποθηκεύει τη συντεταγμένη x για δεδομένο βήμα
x = 0              # Αρχικοποίηση θέσης σωματιδίου
path = [x]

for n in range(N):
    x += step()
    path += [x]

plt.plot(path)     # Βλέπουμε πώς κινείται το σωματίδιο κατά τα Ν βήματα
# Αυξάνοντας τον αριθμό των βημάτων, παρατηρούμε ότι το σωματίδιο επιστρέφει πάντα στην αρχική του θέση, το οποίο είναι κάτι που περιμένουμε, καθώς η αναμενόμενη τιμή της θέσης του είναι μηδέν
```

```python
# Εφαρμογή  Μ = 1000 σωματίδια Brown μετά από Ν = 200 βήματα (εδώ μελετάμε την τελική θέση και όχι τις ενδιάμεσες τιμές)

M = 1000            # Αριθμός σωματιδίων
N = 200             # Αριθμός βημάτων
xArray = []         # Διαστάσεις 1x1000

for m in range(M):
    x = 0
    for n in range(N):
        x += step()
    xArray += [x]
```

```python
xArray[0:10] # Μου δείχνει σε ποιά θέση βρίσκονται τα πρώτα 10 σωματίδια για παράδειγμα μετά τα Ν βήματα
#xArray # Mου δείχνει σε ποιά θέση βρίσκεται το καθένα απ'τα Μ σωμάτια μετά το πέρας Ν βημάτων

plt.hist(xArray, bins = range(-60,61)) # Η κατανομή φαίνεται κάνοντας ένα διάγραμα συχνοτήτων
# Παρατηρούμε ότι αρχίζει και συμπεριφέρεται σαν κανονική κατανομή και στην ουσία δείχνει ότι ξεκινώντας από μια συνάρτηση δελτα (όλα τα σημεία αρχικά βρίσκονται στη θέση x=0) μετά την πάροδο κάποιου χρόνου (200 μοναδιαία βήματα) τα σωματίδια διαχέονται με κανονικό τρόπο γύρω απ' την θέση 0
```

```python
# Προάγοντας τον πίνακα σε numpy array μπορώ να δω τη μέση τιμή του, την τυπική του απόκλιση και  βλέπουμε ότι πράγματι τα σωματίδια βρίσκονται κατανομημένα γύρω απ'τη θέση 0 και πράγματι η τυπική απόκλιση είναι ίση με το sqrt(N), όπως προβλέπεται θεωρητικά.

xArray = np.array(xArray)
print('mean = ', xArray.mean(), 'std = ', xArray.std(), 'sqrt(N) = ', N**0.5)
```
