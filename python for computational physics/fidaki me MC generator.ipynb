```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dice():
    '''returns a random number in [1,6]'''
    U = np.random.rand()
    res = 0
    if   0 < U < 1/6 : res = 1
    if 1/6 < U < 2/6 : res = 2
    if 2/6 < U < 3/6 : res = 3
    if 3/6 < U < 4/6 : res = 4
    if 4/6 < U < 5/6 : res = 5
    if 5/6 < U < 1   : res = 6
    return res
```

```python
array = []
for i in range(0,5000):
    array += [dice()]
```

```python
plt.hist(array, bins = range(1,8))
```

```python
def game(): 
    '''returns number of rolls for winning'''
    ndices = 0 # Πλήθος ζαριών
    pos = 1 # Παρακολουθεί σε ποιά θέση βρίσκεται το πιόνι
    while (pos < 25):
        pos += dice()
        ndices += 1
        if pos == 3 : pos = 24 # σκάλα
        if pos == 21 : pos = 7 # φιδάκι  
        
    return ndices
```

```python
games = [game() for i in range(0,200)] # Δειγματοληψία από αριθμό ζαριών που χρειάστηκαν σε 100 διαφορετικές παρτίδες
#games
plt.hist(games, bins = range(0,20), histtype = 'bar', ec = 'black')
```
