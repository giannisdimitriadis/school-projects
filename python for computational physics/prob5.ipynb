```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Μιγαδικοί στην Python

x = 2
y = 1
z = complex(x,y)
print(z)
print(z.real)
print(z.imag)
```

```python
def step():
    lx = 1
    ly = 0
    l = complex(lx,ly)
    U = np.random.rand()
    if 0.25 < U < 0.50 : l = complex(-1,0)
    if 0.50 < U < 0.75 : l = complex(0,1)
    if 0.75 < U < 1.00 : l = complex(0,-1)
    return l
```

```python
N = 100 # αριθμός βημάτων
x = 0 # Αρχικοποίηση τετμημένης σωματιδίου
y = 0 # Αρχικοποίηση τεταγμένης σωματιδίου
z = complex(x,y) # Συντεταγμένες σωματιδίου
path = [complex(x,y)] # Λίστα που αποθηκεύει τις συντεταγμένες του σωματιδίου στο εκάστοτε βήμα
xpath = [z.real]
ypath = [z.imag]

for n in range(N):
    z += step()
    path += [z]
    xpath += [z.real]
    ypath += [z.imag]


# Βλέπουμε πώς κινείται το σωματίδιο κατά τα Ν βήματα
```

```python
plt.plot(xpath)
plt.plot(ypath)
```

```python
plt.plot(xpath,ypath)
```

```python
M = 1000
N = 100
xArray = [] # Διαστάσεις 1x1000
yArray = [] # Διαστάσεις 1x1000

for m in range(M):
    z = 0
    for n in range(N):
        z += step()
    xArray += [z.real]
    yArray += [z.imag]
```

```python
xArray[0:10] # Βλέπω την απόσταση που απέχουν τα πρώτα 10 σωμάτια απ'τη θέση x=0 στον x άξονα
```

```python
yArray[0:10] # Βλέπω την απόσταση που απέχουν τα πρώτα 10 σωμάτια απ'τη θέση y=0 στον y άξονα
```

```python
# Βλέπουμε ότι η κατανομή της απόστασης που διανύουν στους δύο άξονες τα σωματίδια είναι κανονική 
plt.hist(xArray, bins = range(-60,61))
plt.hist(yArray, bins = range(-60,61))
```

```python
# Προάγοντας τον πίνακα σε numpy array μπορώ να δω τη μέση τιμή του, την τυπική του απόκλιση και  βλέπουμε ότι πράγματι τα σωματίδια βρίσκονται κατανομημένα γύρω απ'τη θέση 0 και πράγματι η τυπική απόκλιση είναι πράγματι ίση με το sqrt(N), όπως προβλέπεται θεωρητικά
xArray = np.array(xArray)
print('mean = ', xArray.mean(), 'std = ', xArray.std())
yArray = np.array(yArray)
print('mean = ', yArray.mean(), 'std = ', yArray.std())
```
